```python
data = pd.read_csv("./train.csv")
```

```python
data.cut.value_counts() # Ordinal value
```

```python
data.color.value_counts() # Thought it was a Nominal value, but it turned out to be ordinal after a research.
```

```python
data.clarity.value_counts() # Ordinal value too
```

```python
color_map = {
    "D": 7,
    "E": 6,
    "F": 5,
    "G": 4,
    "H": 3,
    "I": 2,
    "J": 1
}
```

```python
data["color_num"] = data.color.map(color_map)
data
```

```python
quality_map = {
    "Ideal": 5,
    "Premium": 4,
    "Very Good": 3,
    "Good": 2,
    "Fair": 1
}
```

```python
data["cut_num"] = data.cut.map(quality_map)
data
```

```python
clarity_map = {
    "IF": 8,
    "VVS1": 7,
    "VVS2": 6,
    "VS1": 5,
    "VS2": 4,
    "SI1": 3,
    "SI2": 2,
    "I1": 1
}
```

```python
data["clarity_num"] = data.clarity.map(clarity_map)
data
```

```python
data.drop(["cut", "clarity", "color"], axis=1, inplace=True)
```

```python
data.head()
```

```python
data.to_csv("./train_final.csv")
```
